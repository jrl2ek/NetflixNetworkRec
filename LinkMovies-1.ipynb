```python
def combine(row):
    if row['Name2'] != '':
        row['Name'] = str(row['Name']) + ', ' + str(row['Name2'])
    if row['Name3'] != '':
        row['Name'] = row['Name'] + ', ' + str(row['Name3'])
    if row['Unnamed: 5'] != '':
        row['Unnamed: 5'] = row['Name'] + ', ' + str(row['Unnamed: 5'])
    return row['Name']
```

```python
test = NetflixIMDB.merge(final_movies, left_on='ID', right_on='movieID')
```

```python
test.drop_duplicates(subset=['ID'], inplace=True, keep='first')
```

```python
test.head()
```

```python
test.shape
```

```python
tcnostToTitle = {}
for index, row in test.iterrows():
    tcnostToTitle[row['tconst']] = row['Name']
```

```python
genres = {}
for index, row in test.iterrows():
    #print(row['genres'])
    genreList = str(row['genres']).split(',')
    for genre in genreList:
        if genre not in genres:
            genres[genre] = []
            genres[genre].append((row['Name'],row['ID']))
        else:
            genres[genre].append((row['Name'],row['ID']))
```

```python
len(genres)
```

```python
import networkx as nx
```

```python
for key in genres:
    print(key)
```

```python
genreG = nx.Graph()
count = 0
for key in genres:
    movieList = genres[key].copy()
    #print(movieList)
    
    for i in range(len(movieList)):
        for j in range(i+1,len(movieList)):
            if genreG.has_edge(movieList[i][0], movieList[j][0]):
                #print('in here')
                genreG[movieList[i][0]][movieList[j][0]]['weight'] += 1
            else:
                #print('hi')
                count += 1
                genreG.add_edge(movieList[i][0],movieList[j][0], weight=1)
```

```python
count
```

```python
genreG.number_of_edges()
```

```python
genreG.number_of_nodes()
```

```python
nx.write_pajek(genreG, "GenreG.net")
```
